```python
import pandas as pd
import numpy as np
from sklearn.cluster import SpectralClustering
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
dataset
```

```python
x=dataset.iloc[:,[3,4]].values
cluster= SpectralClustering(assign_labels='discretize', n_clusters=5,
    random_state=0)
labels = cluster.fit_predict(x)
```

```python
  labels
```

```python
spvd=pd.DataFrame(dataset)
spvd['Cluster_group']=labels
spvd
```

```python
import seaborn as sns
```

```python
facet=sns.lmplot(data=spvd,x=spvd.columns[3],y=spvd.columns[4], hue=spvd.columns[5], fit_reg=False, legend=True,legend_out=True)
```
